```python
# listing of correspondence 
n = 0
for l in SeqIO.parse('augustus.whole.aa', 'fasta'):
    for k in SeqIO.parse('peptides.fa', 'fasta'):
        if k.seq in l.seq:
            print('>prot', n)
            print(l.seq, '\n')
            n += 1
```

```python
# parsing of blast output
level = 0.8
n = 0
for l in SearchIO.parse('Z5DFKNDB016-Alignment.xml', 'blast-xml'):
    print(l.description)
    for hsp in l:
        for hs in hsp:
            if 'hypothetical' not in hs.hit_description and 'uncharacterized' not in hs.hit_description\
                    and hs.ident_num / hs.query_span > level:
                print(hs.hit_description)
    print()
```
