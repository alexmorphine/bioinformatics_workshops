```python
from Bio import SeqIO

b = {}
n = 0
for prot in SeqIO.parse('my_result.fasta', 'fasta'):
    if prot.seq not in b:
        b[prot.seq] = 1
    else:
        b[prot.seq] += 1
for i, v in b.items():
    print('>prot', n, 'count', v)
    print(i)
    print()
    n += 1
```

```python
# parsing of BLAST output
level = 0.8
n = 0
for l in SearchIO.parse('Z5DFKNDB016-Alignment.xml', 'blast-xml'):
    print(l.description)
    for hsp in l:
        for hs in hsp:
            if 'hypothetical' not in hs.hit_description and 'uncharacterized' not in hs.hit_description\
                    and hs.ident_num / hs.query_span > level:
                print(hs.hit_description)
    print()
```
